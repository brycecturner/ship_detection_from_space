```python
import os
import sklearn.model_selection  as ms
import shutil
from google.colab import drive
drive.mount('/content/drive')
```

```python
read_base_dir = "drive/MyDrive/ship_detection/yolo_data/images/"
write_base_dir = "drive/MyDrive/ship_detection/yolo_data/test/"
```

```python
all_files = os.listdir(read_base_dir)
train, test = ms.train_test_split(all_files,  test_size=0.25)
```

```python
print(len(train)); print(len(test))
```

```python
for file in test:
    source = read_base_dir + file
    destination = write_base_dir + file
    # move file
    shutil.move(source, destination)
    #print('Moved:', file)
```
